## K-means


```python
import numpy as np

# Um nmero mximo de iteraoes foi estipulado pois quanto maior o nmero, maior o tempo de processamento. E nmeros elevados nao sao passiveis de serem compilados e executados pelo colab.
def k_means(data, k, max_iterations=100):
    # Inicializao dos centroides escolhendo k pontos aleatrios
    centroids = data[np.random.choice(range(len(data)), size=k, replace=False)]

    for iteration in range(max_iterations):
        # Atribuio dos objetos aos grupos
        distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=-1) # Essa linha calcula as distncias entre cada ponto de dados em data e os centroides usando a norma Euclidiana.
        labels = np.argmin(distances, axis=-1) # Aqui os pontos sao atribuidos ao cluster em que se encontram mais perto do seu centroide.

        # Atualizao dos centroides
        new_centroids = np.array([data[labels == i].mean(axis=0) for i in range(k)])

        # Verificao de convergncia porque se os centroides nao mudaram, entao o algoritmo acaba por aqui.
        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    return labels, centroids

```

### O uso com o dataset de heart disease j padronizado, em teste e treino

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split



# Dividir o conjunto de dados em treino e teste
X = df_padronizado.drop('target', axis=1).values
y = df_padronizado['target'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=20, stratify=y)

k = 2  # Numero de clusters
train_labels, centroids = k_means(X_train, k)


# Prever rtulos para os dados de teste
def predict_k_means(data, centroids):
    distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=-1)
    labels = np.argmin(distances, axis=-1)
    return labels

test_labels = predict_k_means(X_test, centroids)

# Para printar os resultados estamos usando PCA
# Reduzindo a dimensao para 2D
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(X_train)
reduced_centroids = pca.transform(centroids)


plt.figure(figsize=(10, 7))

# Plotar pontos de dados
scatter = plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=train_labels, cmap='viridis', marker='o', edgecolor='k')

# Adicionar centroides
plt.scatter(reduced_centroids[:, 0], reduced_centroids[:, 1], c='red', marker='X', s=200, label='Centroids')

# Adicionar a legenda dos clusters
legend1 = plt.legend(*scatter.legend_elements(), title="Clusters")
plt.gca().add_artist(legend1)

plt.xlabel('PCA Componente 1')
plt.ylabel('PCA Componente 2')
plt.title('K-means Clustering of Heart Disease Data - Train')

# Adicionar rtulos aos centroides
for i, centroid in enumerate(reduced_centroids):
    plt.annotate(f'Centroid {i}', (centroid[0], centroid[1]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```

## Separando em teste e treino


```python
# Criar a matriz de confuso
from scipy.stats import mode
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Mapear clusters para rtulos reais
cluster_to_target = {}
for cluster in range(k):
  cluster_mode = mode(y_train[train_labels == cluster])
  cluster_to_target[cluster] = cluster_mode.mode[0] if isinstance(cluster_mode.mode, np.ndarray) else cluster_mode.mode

mapped_train_labels = np.array([cluster_to_target[label] for label in train_labels])
mapped_test_labels = np.array([cluster_to_target[label] for label in test_labels])


# Gerar a matriz de confuso para os dados de treino
cm_train = confusion_matrix(y_train, mapped_train_labels)
ConfusionMatrixDisplay(cm_train, display_labels=['target=0', 'target=1']).plot()
plt.title('Matriz de Confuso - Treino')
plt.show()

# Gerar a matriz de confuso para os dados de teste
cm_test = confusion_matrix(y_test, mapped_test_labels)
ConfusionMatrixDisplay(cm_test, display_labels=['target=0', 'target=1']).plot()
plt.title('Matriz de Confuso - Teste')
plt.show()
```

## rvore de deciso

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns


# Separar x e y
X = df.drop(columns=['target'])
y = df['target']
```

## Tcnica de reamostragem: Stratified K-Fold Cross Validation

O uso de K-fold Cross Validation  crucial em conjuntos de dados desbalanceados para garantir uma avaliao robusta e no enviesada do modelo. Como uma das classes apresenta a proporo de 0.54276315789, mesmo que a diferena  pouca achamos necessrio manter a proporo para que a amostragem de treinamento possa ser o mais parecido possivel com o dataset.

```python
# Dividir os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, stratify=y, random_state=958)

# Exemplo de uso
print("Proporo de pessoas com doena no datset:", y.mean())
print("Proporo de pessoas com doena nos dados de treino:", y_train.mean())
print("Proporo de pessoas com doena nos dados de teste:", y_test.mean())
```

## Parametros utilizados pela arvore

- ### criterion='entropy'
O critrio de entropia  utilizado para medir a impureza de uma diviso em uma rvore de deciso. Em termos de rvores de deciso, uma diviso com menor entropia significa que os ns resultantes so mais homogneos em relao  varivel alvo.



```python
#classificador de rvore de deciso
clf = DecisionTreeClassifier(criterion='entropy',random_state=958)


clf.fit(X_train, y_train)
```

```python
plt.figure(figsize=(66, 15))
plot_tree(clf, filled=True, feature_names=X.columns, class_names=['sem doena', 'com doena'],  fontsize=9, precision=2)
plt.title("rvore de Deciso", fontsize=15)  # Ttulo da rvore
plt.savefig('arvore_decisao.png', format='png', dpi=300)
plt.show()
```

```python
# Previses nos dados de treino e teste
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)


print(f'Acurcia nos dados de treino: {accuracy_score(y_train, y_train_pred)}')
print(f'Acurcia nos dados de teste:  {accuracy_score(y_test, y_test_pred)}')
```

```python
#Funo para plotar a matrz de confuso
def plotar_matrix(y_real, y_previsto, title):

    # Calcular
    cm = confusion_matrix(y_real, y_previsto)

    # Plotar
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, cmap='Greens', fmt='d', cbar=False,
                annot_kws={'fontsize': 18},
                xticklabels=['Sem Doena', 'Com Doena'],
                yticklabels=['Sem Doena', 'Com Doena'])
    plt.xlabel('Valor Previsto')
    plt.ylabel('Valor Real')
    plt.title(title)
    plt.savefig(title+'.png', format='png', dpi=400)
    plt.show()

```

```python

plotar_matrix(y_train, y_train_pred, 'Matriz de Confuso - Dados de Treinamento')
plotar_matrix(y_test, y_test_pred, 'Matriz de Confuso - Dados de Teste')
```

# Deteco de overfitting
O modelo apresentado teve um acerto de 100% no conjunto de treinamento, significando que o modelo apresenta overfitting.
Para resolver iremos refazer o modelo, mas limitando a profundidade da rvore.

- ### max_depth
Limitar a profundidade da rvore  uma tcnica de regularizao que visa prevenir o overfitting.

## Modelo ajustado limitando a profundidade

Primeiramente, vamos determinar qual o melhor valor para o max_depth

```python


melhor_max_depth = 0
melhor_accuracy = 0


for max_depth in range(1, 11):
    clf = DecisionTreeClassifier(criterion='entropy', random_state=958, max_depth=max_depth)
    clf.fit(X_train, y_train)
    y_test_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_test_pred)
    if accuracy > melhor_accuracy:
            melhor_accuracy = accuracy
            melhor_max_depth = max_depth


print(f'Melhor Max Depth: {melhor_max_depth}')
```

## Melhor Max Depth: 3

```python
#classificador de rvore de deciso
clf = DecisionTreeClassifier(criterion='entropy', max_depth=3,random_state=958)


clf.fit(X_train, y_train)
plt.figure(figsize=(46, 10))
plot_tree(clf, filled=True, feature_names=X.columns, class_names=['sem doena', 'com doena'],  fontsize=9, precision=2)
plt.title("rvore de Deciso", fontsize=15)  # Ttulo da rvore
plt.savefig('arvore_decisao.png', format='png', dpi=300)
plt.show()

y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)


print(f'Acurcia nos dados de treino: {accuracy_score(y_train, y_train_pred)}')
print(f'Acurcia nos dados de teste:  {accuracy_score(y_test, y_test_pred)}')
plotar_matrix(y_train, y_train_pred, 'Matriz de Confuso - Dados de Treinamento')
plotar_matrix(y_test, y_test_pred, 'Matriz de Confuso - Dados de Teste')
```
