## Por que o uso de dados padronizados?

Padronizar os dados para aplicação no algoritmo K-means é fundamental para assegurar que todas as variáveis tenham média zero e desvio padrão unitário. Isso é crucial porque o K-means utiliza distâncias euclidianas para determinar a similaridade entre pontos de dados e centróides de cluster.

## Dados padronizados


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_padronizado = df.copy()
for atributo, tipo in zip(df_padronizado.columns, df_padronizado.dtypes):

    if (tipo == 'int64' or atributo =='float64') and atributo !='target':

        df_padronizado[atributo] = scaler.fit_transform(df_padronizado[[atributo]])
    else:
        df_padronizado[atributo] = df_padronizado[atributo]

df_padronizado.head()
```

## K-means


```python
import numpy as np

# Um número máximo de iteraçoes foi estipulado pois quanto maior o número, maior o tempo de processamento. E números elevados nao sao passiveis de serem compilados e executados pelo colab.
def k_means(data, k, max_iterations=100):
    # Inicialização dos centroides escolhendo k pontos aleatórios
    centroids = data[np.random.choice(range(len(data)), size=k, replace=False)]

    for iteration in range(max_iterations):
        # Atribuição dos objetos aos grupos
        distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=-1) # Essa linha calcula as distâncias entre cada ponto de dados em data e os centroides usando a norma Euclidiana.
        labels = np.argmin(distances, axis=-1) # Aqui os pontos sao atribuidos ao cluster em que se encontram mais perto do seu centroide.

        # Atualização dos centroides
        new_centroids = np.array([data[labels == i].mean(axis=0) for i in range(k)])

        # Verificação de convergência porque se os centroides nao mudaram, entao o algoritmo acaba por aqui.
        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    return labels, centroids

```

### O uso com o dataset de heart disease já padronizado, em teste e treino

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split



# Dividir o conjunto de dados em treino e teste
X = df_padronizado.drop('target', axis=1).values
y = df_padronizado['target'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=20, stratify=y)

k = 2  # Numero de clusters
train_labels, centroids = k_means(X_train, k)


# Prever rótulos para os dados de teste
def predict_k_means(data, centroids):
    distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=-1)
    labels = np.argmin(distances, axis=-1)
    return labels

test_labels = predict_k_means(X_test, centroids)

# Para printar os resultados estamos usando PCA
# Reduzindo a dimensao para 2D
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(X_train)
reduced_centroids = pca.transform(centroids)


plt.figure(figsize=(10, 7))

# Plotar pontos de dados
scatter = plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=train_labels, cmap='viridis', marker='o', edgecolor='k')

# Adicionar centroides
plt.scatter(reduced_centroids[:, 0], reduced_centroids[:, 1], c='red', marker='X', s=200, label='Centroids')

# Adicionar a legenda dos clusters
legend1 = plt.legend(*scatter.legend_elements(), title="Clusters")
plt.gca().add_artist(legend1)

plt.xlabel('PCA Componente 1')
plt.ylabel('PCA Componente 2')
plt.title('K-means Clustering of Heart Disease Data - Train')

# Adicionar rótulos aos centroides
for i, centroid in enumerate(reduced_centroids):
    plt.annotate(f'Centroid {i}', (centroid[0], centroid[1]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```

## Separando em teste e treino


```python
# Criar a matriz de confusão
from scipy.stats import mode
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Mapear clusters para rótulos reais
cluster_to_target = {}
for cluster in range(k):
  cluster_mode = mode(y_train[train_labels == cluster])
  cluster_to_target[cluster] = cluster_mode.mode[0] if isinstance(cluster_mode.mode, np.ndarray) else cluster_mode.mode

mapped_train_labels = np.array([cluster_to_target[label] for label in train_labels])
mapped_test_labels = np.array([cluster_to_target[label] for label in test_labels])


# Gerar a matriz de confusão para os dados de treino
cm_train = confusion_matrix(y_train, mapped_train_labels)
ConfusionMatrixDisplay(cm_train, display_labels=['target=0', 'target=1']).plot()
plt.title('Matriz de Confusão - Treino')
plt.show()

# Gerar a matriz de confusão para os dados de teste
cm_test = confusion_matrix(y_test, mapped_test_labels)
ConfusionMatrixDisplay(cm_test, display_labels=['target=0', 'target=1']).plot()
plt.title('Matriz de Confusão - Teste')
plt.show()
```

## Árvore de decisão

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns


# Separar x e y
X = df.drop(columns=['target'])
y = df['target']
```

## Técnica de reamostragem: Stratified K-Fold Cross Validation

O uso de K-fold Cross Validation é crucial em conjuntos de dados desbalanceados para garantir uma avaliação robusta e não enviesada do modelo. Como uma das classes apresenta a proporção de 0.54276315789, mesmo que a diferença é pouca achamos necessário manter a proporção para que a amostragem de treinamento possa ser o mais parecido possivel com o dataset.

```python
# Dividir os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, stratify=y, random_state=958)

# Exemplo de uso
print("Proporção de pessoas com doença no datset:", y.mean())
print("Proporção de pessoas com doença nos dados de treino:", y_train.mean())
print("Proporção de pessoas com doença nos dados de teste:", y_test.mean())
```

## Parametros utilizados pela arvore

- ### criterion='entropy'
O critério de entropia é utilizado para medir a impureza de uma divisão em uma árvore de decisão. Em termos de árvores de decisão, uma divisão com menor entropia significa que os nós resultantes são mais homogêneos em relação à variável alvo.



```python
#classificador de árvore de decisão
clf = DecisionTreeClassifier(criterion='entropy',random_state=958)


clf.fit(X_train, y_train)
```

```python
plt.figure(figsize=(66, 15))
plot_tree(clf, filled=True, feature_names=X.columns, class_names=['sem doença', 'com doença'],  fontsize=9, precision=2)
plt.title("Árvore de Decisão", fontsize=15)  # Título da árvore
plt.savefig('arvore_decisao1.png', format='png', dpi=300)
plt.show()
```

```python
# Previsões nos dados de treino e teste
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)


print(f'Acurácia nos dados de treino: {accuracy_score(y_train, y_train_pred)}')
print(f'Acurácia nos dados de teste:  {accuracy_score(y_test, y_test_pred)}')
```

```python
#Função para plotar a matrz de confusão
def plotar_matrix(y_real, y_previsto, title):

    # Calcular
    cm = confusion_matrix(y_real, y_previsto)

    # Plotar
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, cmap='Greens', fmt='d', cbar=False,
                annot_kws={'fontsize': 18},
                xticklabels=['Sem Doença', 'Com Doença'],
                yticklabels=['Sem Doença', 'Com Doença'])
    plt.xlabel('Valor Previsto')
    plt.ylabel('Valor Real')
    plt.title(title)
    plt.savefig(title+'.png', format='png', dpi=400)
    plt.show()

```

```python

plotar_matrix(y_train, y_train_pred, 'Matriz de Confusão - Dados de Treinamento 1')
plotar_matrix(y_test, y_test_pred, 'Matriz de Confusão - Dados de Teste 1')
```

# Detecção de overfitting
O modelo apresentado teve um acerto de 100% no conjunto de treinamento, significando que o modelo apresenta overfitting.
Para resolver iremos refazer o modelo, mas limitando a profundidade da árvore.

- ### max_depth
Limitar a profundidade da árvore é uma técnica de regularização que visa prevenir o overfitting.

## Modelo ajustado limitando a profundidade

Primeiramente, vamos determinar qual o melhor valor para o max_depth

```python


melhor_max_depth = 0
melhor_accuracy = 0
acuracias = []

for max_depth in range(1, 11):
    clf = DecisionTreeClassifier(criterion='entropy', random_state=958, max_depth=max_depth)
    clf.fit(X_train, y_train)
    y_test_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_test_pred)
    acuracias.append(accuracy)
    
    if accuracy > melhor_accuracy:
        melhor_accuracy = accuracy
        melhor_max_depth = max_depth

print(f'Melhor Max Depth: {melhor_max_depth}')

# Plotar as acurácias
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), acuracias, marker='o', linestyle='-', color='b')
plt.title('Acurácia por Profundidade Máxima da Árvore de Decisão')
plt.xlabel('Profundidade Máxima')
plt.ylabel('Acurácia')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
plt.savefig('acuracia_por_profundidade_maxima.png')
```

## Melhor Max Depth: 3
